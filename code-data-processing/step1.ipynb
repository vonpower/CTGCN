```python
import pandas as pd
import numpy as np
```

```python
# 非智能合约所有交易记录
transaction = pd.read_excel("transaction.xlsx")
transaction
```

```python
# 实验所用的交易记录在30-100的账户信息
address = pd.read_excel("address.xlsx")
address
```

```python
froms = list(transaction['Froms'])
tos = list(transaction['Tos'])
account = list(address['adress'])
account[1]
```

```python
froms_new = [i.lower() for i in froms]
tos_new = [i.lower() for i in tos]
account_new = [i.lower() for i in account]
transaction['froms_new'] = froms_new
transaction['tos_new'] = tos_new
transaction
```

```python
# 在所有交易记录中提取记录为30-100的用户交易信息
from_trans = transaction.loc[transaction['froms_new'].isin(account_new)]
to_trans = transaction.loc[transaction['tos_new'].isin(account_new)]
```

```python
type(from_trans)
```

```python
data = from_trans.append(to_trans)
data
```

```python
data.loc[data['to_balances']=='None', 'to_balances'] = 0
data.loc[data['from_balances']=='None', 'from_balances'] = 0
```

```python
data.to_excel('data.xlsx')
```
