## 图ID与节点ID对应

```python
from_ad = transaction['from_ad']
to_ad = transaction['to_ad']
from_id = transaction['from_id']
to_id = transaction['to_id']
```

```python
data = pd.DataFrame()
data['from_ad'] = from_ad
data['to_ad'] = to_ad
data['from_id'] = from_id
data['to_id'] = to_id
data
```

```python
from_graph_id = []
for i in from_ad:
    if i in ad:
        index = ad.index(i)
        from_graph_id.append(graph_id[index])
    else:
        from_graph_id.append(0)
len(from_graph_id)
```

```python
to_graph_id = []
for i in to_ad:
    if i in ad:
        index = ad.index(i)
        to_graph_id.append(graph_id[index])
    else:
        to_graph_id.append(0)
len(to_graph_id)
```

```python
data['from_graph_id'] = from_graph_id
data['to_graph_id'] = to_graph_id
data
```

### 按照graph_id把0替换为该图ID

```python
for i in from_graph_id:
    if i == 0:
        index = from_graph_id.index(i)
        from_graph_id[index] = to_graph_id[index]
for i in to_graph_id:
    if i == 0:
        index = to_graph_id.index(i)
        to_graph_id[index] = from_graph_id[index]
```

```python
data['from_graph_id'] = from_graph_id
data['to_graph_id'] = to_graph_id
data
```

```python
## 节点地址
node_id = list(from_id)
node_id.extend(list(to_id))
## 节点标签
graph1_id = list(from_graph_id)
graph1_id.extend(list(to_graph_id))
## 合并节点和标签
node_graph = pd.DataFrame(zip(node_id,graph1_id))
node_graph.columns = ['node_id','graph1_id']
node_graph
```

```python
# 删除重复值
node_graph_new = node_graph.drop_duplicates(subset=['node_id','graph1_id'])
node_graph_new
```

```python
grouped = node_graph_new.groupby('node_id')
```
