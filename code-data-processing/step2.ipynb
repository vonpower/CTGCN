## 导入原始数据

```python
import pandas as pd
import numpy as np
```

```python
# 交易记录
transaction = pd.read_excel("data.xlsx")
transaction
```

```python
gas = transaction['gas']
gasPrice = transaction['gasPrice']
```

```python
# 根据gas*gasprice得到交易手续费，即trans_fee
trans_fee = []
for i in range(len(transaction['Froms'])):
    trans_fee.append(gas[i]*gasPrice[i])
len(trans_fee)
```

```python
transaction['trans_fee'] = trans_fee
df = transaction
df
```

## 统计两个相同之间的交易次数和金额，交易时间

#### 交易金额、交易次数求和
#### 求交易时间间隔

```python
froms = df['Froms']
tos = df['Tos']
```

```python
froms = [i.lower() for i in froms]
tos = [i.lower() for i in tos]
```

```python
# 所有交易信息
lista = []
for i in range(len(froms)):
    lista.append(froms[i]+tos[i])
print(lista[0])
len(lista)
```

```python
df['from_to'] = lista
df
```

```python
grouped = df.groupby('from_to')
```

```python
address = list(grouped.groups)# 获取分组地址
value = list(grouped['Amount'].sum())# 交易金额求和
time_inter = list(grouped['blockNumber'].max()-grouped['blockNumber'].min())# 时间间隔
fee = list(grouped['trans_fee'].sum())# 交易费用求和
count = list(grouped['from_to'].count())# 交易次数
from_balance = list(grouped['from_balances'].sum()) # 求余额之和
to_balance = list(grouped['to_balances'].sum())
```

```python
from_ad = []
to_ad = []
for ad in address:
    from_ad.append(ad[0:42])
    to_ad.append(ad[42:])
```

```python
data_final = pd.DataFrame()
data_final['from_ad'] = from_ad
data_final['to_ad'] = to_ad
data_final['value'] = value
data_final['time_inter'] = time_inter
data_final['fee'] = fee
data_final['count'] = count
data_final['from_balance'] = from_balance
data_final['to_balance'] = to_balance
data_final
```

## 增加标签

```python
account = pd.read_excel("address.xlsx")
account
```

```python
adress = list(account['adress'])
adress = [i.lower() for i in adress]
label = list(account['label'])
```

```python
from_label = []
for ad in data_final['from_ad']:
    if ad in adress:
        index = adress.index(ad)
        from_label.append(label[index])
    else:
        from_label.append(2)
print(len(from_label))
```

```python
to_label = []
for ad in data_final['to_ad']:
    if ad in adress:
        index = adress.index(ad)
        to_label.append(label[index])
    else:
        to_label.append(2)
print(len(to_label))
```

```python
data_final['from_label'] = from_label
data_final['to_label'] = to_label
data_final
```

## 计算平均余额：余额之和/该用户总的交易次数

```python
grouped1 = data_final.groupby('from_ad')
address1 = list(grouped1.groups)
count1 = list(grouped1['count'].sum())# 交易次数
from_balance1 = list(grouped1['from_balance'].sum())# 账户余额
from_count = []
from_sum_balance = []
for ad in data_final['from_ad']:
    index = address1.index(ad)
    from_sum_balance.append(from_balance1[index])
    from_count.append(count1[index])
from_avg_balance = list(map(lambda x , y : x / y, from_sum_balance,from_count))
# from_avg_balance
```

```python
grouped2 = data_final.groupby('to_ad')
address2 = list(grouped2.groups)
count2 = list(grouped2['count'].sum())# 交易次数
to_balance1 = list(grouped1['to_balance'].sum())# 账户余额
to_count = []
to_sum_balance = []
for ad in data_final['to_ad']:
    index = address2.index(ad)
    to_sum_balance.append(to_balance1[index])
    to_count.append(count2[index])
to_avg_balance = list(map(lambda x , y : x / y, to_sum_balance,to_count))
```

```python
data_final['from_avg_balance'] = from_avg_balance
data_final['to_avg_balance'] = to_avg_balance
data_final['from_count'] = from_count
data_final['to_count'] = to_count
data_final
```

## 计算权重：交易次数/该用户总的交易次数

```python
weigth = []
for i in range(len(to_count)):
    if from_label[i] == 2:
        weigth.append(count[i]/to_count[i])
    if to_label[i] == 2:
        weigth.append(count[i]/from_count[i])
    if from_label[i] != 2 and to_label[i] != 2:
        weigth.append(count[i]/(from_count[i]+to_count[i]))
len(weigth)
```

```python
data_final['weight'] = weigth
data_final
```

## 生成ID

### 节点ID

```python
## 节点地址
address = list(from_ad)
address.extend(list(to_ad))
## 节点标签
label = list(from_label)
label.extend(list(to_label))
## 合并节点和标签
nodes = pd.DataFrame(zip(address,label))
nodes.columns = ['address','label']
nodes
```

```python
nodes_new = nodes.drop_duplicates(subset=['address','label'])
nodes_new
```

```python
import numpy as np
node_id = list(np.arange(1,len(nodes_new)+1))
nodes_new['node_ID'] = node_id
nodes_new
```

```python
nodes_new.to_excel('data_nodes.xlsx')
```

```python
node_address = list(nodes_new['address'])
node_ID = list(nodes_new['node_ID'])
```

```python
b = dict(zip(node_address,node_ID))
data_final['from_id']=data_final['from_ad'].map(b) 
data_final['to_id']=data_final['to_ad'].map(b) 
data_final
```
